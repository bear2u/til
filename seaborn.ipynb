````
import pandas as pd
import seaborn as sns
````

````
df = pd.read_csv('https://raw.githubusercontent.com/bear2u/datascience-study2/master/csv/subway.csv')

# 승차인원에 대한 내용
sns.kdeplot(df['in'])
````

````
df = pd.read_csv('https://raw.githubusercontent.com/bear2u/datascience-study2/master/csv/salary.md')
sns.violinplot(df['salary'])
````

````
df = pd.read_csv('https://raw.githubusercontent.com/bear2u/datascience-study2/master/csv/laptops.csv')
# df
# df['os'].unique()

# 카테고리화
sns.catplot(data=df, x='os', y='price', kind='box')
````

````
sns.catplot(data=df, x='os', y='price', kind='strip')
````

````
sns.catplot(data=df, x='os', y='price', kind='strip', hue='processor_brand')
````

````
# 좀 더 펼쳐지게 구성
sns.catplot(data=df, x='os', y='price', kind='swarm', hue='processor_brand')
````
